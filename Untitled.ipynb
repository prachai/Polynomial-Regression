```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x=2-3*np.random.normal(0,1,20)
y=x-2*(x**2)+0.5*(x**3)+np.random.normal(-3,3,20)
```

```python
model=LinearRegression()
```

```python
model.fit(x.reshape(-1,1),y.reshape(-1,1))
```

```python
y_pred=model.predict(x.reshape(-1,1))
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,label='linear regression')
plt.show()
```

```python
import sklearn.metrics as metrics
```

```python
metrics.mean_squared_error(y,y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y,y_pred))
```

```python
metrics.r2_score(y,y_pred)
```

# POLYNOMIAL REGRESSION

```python
np.random.seed(0)
x=2-3*np.random.normal(0,1,20)
y=x-2*(x**2)+0.5*(x**3)+np.random.normal(-3,3,20)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features=PolynomialFeatures(degree=3)
```

```python
x_poly=poly_features.fit_transform(x.reshape(-1,1))
```

```python
mod1=LinearRegression()
```

```python
mod1.fit(x_poly,y)
```

```python
y_poly=mod1.predict(x_poly)
```

```python
np.sqrt(metrics.mean_squared_error(y,y_poly))
```

```python
metrics.r2_score(y,y_poly)
```

```python
plt.scatter(x,y)
import operator
sort_axis=operator.itemgetter(0)
sort_zip=sorted(zip(x,y_poly),key=sort_axis)
x,y_poly=zip(*sort_zip)
#plt.plot(x,y_pred,label='Linear')
plt.plot(x,y_poly,label='Polynomial')
plt.xlabel('predict')
plt.ylabel('target')
plt.show()
```

#### From the below picture we can observe that as the model complexity increases, the bias decreases and the variance increases and vice-versa. Ideally, a machine learning model should have low variance and low bias. But practically its impossible to have both. Therefore to achieve a good model that performs well both on the train and unseen data, a trade-off is made.
